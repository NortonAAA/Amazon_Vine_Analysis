```python
# Dependencies and Setup
import pandas as pd
import numpy as np

#File to Laod
vine_table_to_load = "Resources/vine_table.csv"
#Read the table
vine_table_df = pd.read_csv(vine_table_to_load)
```

```python
#Check Vine_Table
vine_table_df.head(10)
```

```python
# Step 1: Filter by Total Votes greater than or equal to 20
vine_table_filtered = vine_table_df[(vine_table_df["total_votes"] >= 20)]
vine_table_filtered.head(10)
```

```python
# Step 2
# Filter the new DataFrame or table created in Step 1 and create a new DataFrame or table to retrieve 
# all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%

vine_table_helpful = vine_table_filtered[((vine_table_filtered["helpful_votes"] / vine_table_filtered["total_votes"]) >= 0.5)]
vine_table_helpful.head(10)
```

```python
# Step 3 - Filter where Vine was Paid
vine_table_paid = vine_table_helpful[(vine_table_helpful["vine"] == "Y")]
vine_table_paid.head()
```

```python
# Step 4 - Filter where Vine was Unpaid
vine_table_unpaid = vine_table_helpful[(vine_table_helpful["vine"] == "N")]
vine_table_unpaid.head()
```

```python
#Step 5a - Total Number of Reviews, 5-Star Reviews, % of 5 Stars for Paid
Total_reviews_paid = len(vine_table_paid["review_id"])
print(f"Total Paid Reviews: {Total_reviews_paid}")
Total_reviews_unpaid = len(vine_table_unpaid["review_id"])
print(f"Total Unpaid Reviews: {Total_reviews_unpaid}")
```

```python
#Get 5 Star Reviews by Type
Paid_five_star = vine_table_paid.loc[vine_table_paid["star_rating"]==5]["review_id"].count()
print(f"Total Paid Five Star Reviews: {Paid_five_star}")

Unpaid_five_star = vine_table_unpaid.loc[vine_table_unpaid["star_rating"]==5]["review_id"].count()
print(f"Total Unpaid Five Star Reviews: {Unpaid_five_star}")
```

```python
#Get Percentage by Type
Per_Paid_five_star = float(Paid_five_star / Total_reviews_paid *100)
print(f"Total Paid % Five Star Reviews: {round(Per_Paid_five_star)}%")
Per_Unpaid_five_star = float(Unpaid_five_star / Total_reviews_unpaid *100)
print(f"Total Unpaid % Five Star Reviews: {round(Per_Unpaid_five_star)}%")
```

```python
#Build into two dataframes by Type
Paid_df = {"Type":"Paid","Total Reviews":Total_reviews_paid,"5 Star Reviews":Paid_five_star,"% 5 Stars":Per_Paid_five_star}
Unpaid_df = {"Type":"Unpaid","Total Reviews":Total_reviews_unpaid,"5 Star Reviews":Unpaid_five_star,"% 5 Stars":Per_Unpaid_five_star}
```

```python
#Merge Type Data Frames into 1
Summary_df = pd.DataFrame([Paid_df,Unpaid_df])
Summary_df
```

```python
#Format columns for summary ReadMe
Summary_df["Total Reviews"] = Summary_df["Total Reviews"].map("{:,.0f}".format)
Summary_df["5 Star Reviews"] = Summary_df["5 Star Reviews"].map("{:,.0f}".format)
Summary_df["% 5 Stars"] = Summary_df["% 5 Stars"].map("{:,.1f}%".format)
Summary_df
```
