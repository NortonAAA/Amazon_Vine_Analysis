```python
# Dependencies and Setup
import pandas as pd
import numpy as np

#File to Laod
vine_table_to_load = "Resources/vine_table.csv"
#Read the table
vine_table_df = pd.read_csv(vine_table_to_load)
```

```python
#Check Vine_Table
vine_table_df.head(10)
```

```python
# Step 1: Filter by Total Votes greater than or equal to 20
vine_table_filtered = vine_table_df[(vine_table_df["total_votes"] >= 20)]
vine_table_filtered.head(10)
```

```python
# Step 2
# Filter the new DataFrame or table created in Step 1 and create a new DataFrame or table to retrieve 
# all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%

vine_table_helpful = vine_table_filtered[((vine_table_filtered["helpful_votes"] / vine_table_filtered["total_votes"]) >= 0.5)]
vine_table_helpful.head(10)
```

```python
# Step 3 - Filter where Vine was Paid
vine_table_paid = vine_table_helpful[(vine_table_helpful["vine"] == "Y")]
vine_table_paid.head()
```

```python
# Step 4 - Filter where Vine was Unpaid
vine_table_unpaid = vine_table_helpful[(vine_table_helpful["vine"] == "N")]
vine_table_unpaid.head()
```
